# Functions

```python
import numpy as np
```

```python
radius = []
for i in range(1,5):
    radius.append(i)
radius
```

```python
circum=[]
for i in radius:
    #print (2*np.pi*i)
    circum.append(2*np.pi*i)
    
```

```python
circum
```

```python
def circumf(radius):
    return np.round(2*np.pi*radius,2),np.round(np.pi*radius**2,2)
```

```python
circ,area=circumf(3)
print (circ)
print (area)
```

```python
circumf(3)[1]
```

### List comprehension

```python
radius = [2*i for i in range(1,5)]
radius
```

```python
# To find circumference
print ("circumference:",[circumf(i)[0] for i in radius])

# To find area
[circumf(i)[1] for i in radius]
```

# Visualization Using Seaborn

```python
#https://seaborn.pydata.org/
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
titanic=pd.read_csv('data/titanic_data.csv')
# titanic dataframe
```

```python
titanic.head()
```

```python
titanic.dtypes
```

```python
#Countplot
# Used when one of the vairable is categorical
```

```python
sns.countplot(y='Sex',data=titanic);
```

```python
sns.countplot(y='Sex',data=titanic, hue='Pclass');
```

```python
# Scatterplot
# Both data numeric
sns.scatterplot(x='Age',y='Fare',data=titanic);
```

```python
sns.scatterplot(x='Age',y='Fare',data=titanic,hue='Sex');
```

### Relational Plots

```python
# Continous data, integer
sns.relplot(x='Age',y='Fare',data=titanic,kind='scatter', hue='Sex');
```

```python
sns.relplot(x='Age',y='Fare',data=titanic,kind='scatter',col='Pclass',hue='Sex');
```

```python
sns.relplot(x='Age',y='Fare',data=titanic,kind='scatter',row='Pclass',hue='Sex');
```

```python
sns.relplot(x='Age',y='Fare',data=titanic,kind='scatter',
            style='Pclass',aspect=2,hue='Sex');
```

```python
# Lineplot
sns.relplot(x='Age',y='Fare',data=titanic,kind='line',hue='Pclass',palette=['b','r','g'],ci=None);
```

```python
# https://seaborn.pydata.org/generated/seaborn.relplot.html
```

# Categorical Plots

```python
#Categorical data
sns.catplot(x='Sex',kind='count',data=titanic);
```

```python
# https://seaborn.pydata.org/tutorial/categorical.html#:~:text=In%20general%2C%20the%20seaborn%20categorical%20plotting%20functions%20try,axis%20looks%20numerical%2C%20the%20levels%20will%20be%20sorted.
```

```python
sns.catplot(x='Sex',kind='count',hue='Survived',data=titanic);
```

```python
sns.catplot(x='Sex',y='Age',kind='bar',hue='Survived',data=titanic,ci=None,order=['female','male']);
```

```python
sns.catplot(x='Pclass',y='Age',kind='point',data=titanic);
```

```python
sns.catplot(x='Pclass',y='Age',kind='point',data=titanic,join=False,hue='Sex');
```

```python
sns.catplot(x='Pclass',y='Age',kind='point',data=titanic,join=False,hue='Sex',ci=None);
```

```python
# Changing plot style
# sns.set_style('whitegrid')
sns.catplot(x='Pclass',y='Age',kind='point',data=titanic,join=False,hue='Sex',ci=None);
```

```python
sns.catplot(x='Pclass',y='Age',kind='point',data=titanic,join=False,hue='Sex',palette='Purples');
```

```python
sns.set_palette("RdBu")
sns.catplot(x='Pclass', y="Age", kind ='bar', data=titanic)
plt.savefig('class2.png')
```
